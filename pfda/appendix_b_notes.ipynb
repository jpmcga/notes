```python
# Timing code with %time and %timeit
# %time runs code ones and records the time
a = 'foobar'
b = 'foo'

%time a.startswith(b)
%time a[:3] == b
```

```python
# %timeit for many repeats
%timeit a.startswith(b)
%timeit a[:3] == b
```

B.4 Advanced IPython Features

```python
# Getting classes to display pretty in IPython with __repr__
class Message:
    def __init__(self, msg):
        self.msg = msg
        
    def __repr__(self):
        return 'Message: %s' % self.msg

x = Message('blahblahblah') 
x
```
