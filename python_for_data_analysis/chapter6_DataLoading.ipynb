JSON

```python
import json
import pandas as pd

test_json = """
{"name" : "James",
"some traits" : ["I", "am", "tall"],
"age" : 30
}
"""
# json.loads convert json into string
jsn = json.loads(test_json)
print(type(jsn))
print(jsn["age"])
# json.dumps converts back to json
jsn2 = json.dumps(jsn) 
print(jsn2)
print(type(jsn2))

# pandas can read a jsn from a dict
df1 = pd.DataFrame(jsn)
print(df1)
# or directly from json
df2 = pd.read_json(test_json)
print(df2)
```

SQL databases

```python
import sqlite3
```

```python
# Creating a table
query = """
CREATE TABLE test
    (name VARCHAR(20),
    age INTEGER,
    address VARCHAR(20)
    );
"""

connection = sqlite3.connect('mydata.sqlite')
connection.execute(query)


```

```python
# Inserting data
data = [('Bob', '34', 'Elliot St'),
        ('Zoe', '34', 'Clark Street'),
        ('Wendy', '48', 'Fellows Balls')
       ]

stmt = "INSERT INTO test VALUES(?, ?, ?)"
connection.executemany(stmt, data)
connection.commit()
```

```python
# Getting back data
x = connection.execute('SELECT * FROM test')
rows = x.fetchall()
x.description
pd.DataFrame(rows, columns=[x[0] for x in x.description])
```
